### Importing Modules and importing data 

```python
import statistics as st
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
os.chdir(r'/home/hrv/Downloads')
df = pd.read_csv("exodata.csv")
```

### Mean, Median, Mode and Standard Deviation of log(g) values of host stars

```python
logg = np.array(df.st_logg)
logg = logg[~(np.isnan(logg))]

mean = np.mean(logg)
median = np.median(logg)
mode = st.mode(logg)
sd = np.std(logg)
print("Mean of log g of stars: %.3f"%mean)
print("Median of log g of stars: %.3f"%median)
print("Mode of log g of stars: %.3f"%mode)
print("Standard Deviation: %.3f"%sd)
```

### Plot of log(g) Values

```python
G = 6.67e-8
M = 2e33
R = 6.96e10
g_sun = (G*M)/(R**2)
logg_sun = np.log10(g_sun)

plt.plot(logg,'o',markersize=2)
plt.axhline(mean,color='r',linestyle='--',label='Mean')
plt.axhline(logg_sun,color='y',linestyle='--',label='Solar log(g)')
plt.axhline(mean+sd,color='k',linestyle='-.',label=r'1$\sigma$')
plt.axhline(mean-sd,color='k',linestyle='-.')
plt.axhline(mean+(2*sd),color='g',linestyle='-.',label=r'2$\sigma$')
plt.axhline(mean-(2*sd),color='g',linestyle='-.')
plt.ylabel(r'log(g) [log10(cm/s$^2$)]')
plt.legend(loc='best')
plt.show()
```

### Histogram of log(g)

```python
plt.hist(logg,bins=50,edgecolor='k')
plt.xlabel(r'log(g) value of host stars [log10(cm/s$^2$)]')
plt.ylabel('Frequency')
plt.show()
```

- From the above histogram, we can see that most of the stars have a log(g) value in the range 4 to 5. The mean of the distribution is close to the log(g) value of stars which means most stars have the same log(g) value as that of the Sun. There is a slight central tendency in the distribution, which is also evident from the fact that the mean and mode of the data are almost same.  The distribution is not symmetric but is slighty negatively skewed.

### Mean, Median, Mode and Standard Deviation of Effective Temperature of host stars

```python
teff = np.array(df.st_teff)
teff = teff[~(np.isnan(teff))]

mean = np.mean(teff)
median = np.median(teff)
mode = st.mode(teff)
sd = np.std(teff)
print("Mean effective temperature of stars: %.3f"%mean)
print("Median effective temperature of stars: %.3f"%median)
print("Mode of effective temperatue of stars: %.3f"%mode)
print("Standard Deviation: %.3f"%sd)
```

### Plot of Effective Temperature

```python
teff_sun = 5780

plt.plot(teff,'o',markersize=2)
plt.axhline(mean,color='r',linestyle='--',label='Mean')
plt.axhline(teff_sun,color='y',linestyle='--',label='Effective temperature of Sun')
plt.axhline(mean+sd,color='k',linestyle='-.',label=r'1$\sigma$')
plt.axhline(mean-sd,color='k',linestyle='-.')
plt.axhline(mean+(2*sd),color='g',linestyle='-.',label=r'2$\sigma$')
plt.axhline(mean-(2*sd),color='g',linestyle='-.')
plt.ylabel('Effective Temperature (K)')
plt.legend(loc='best')
plt.show()
```

### Histogram of Effective Temperature

```python
plt.hist(teff,bins=50,edgecolor='k')
plt.xlabel('Effective Temperature of host stars (K)')
plt.ylabel('Frequency')
plt.show()
```

- From the above histogram, we can see that most of the stars have an effective temperature in the range 5000-6000 K. The mean of this distribution is also cose to the effective temperature of Sun which again shows that most of the planet hosting stars are Sun-like. We also observe that there are very few stars with effective temperature much higher than that of the Sun. This distribution is also not perfectly symmetric.

### Mean, Median, Mode and Standard Deviation of Age of host stars

```python
age = np.array(df.st_age)
age = age[~(np.isnan(age))]

mean = np.mean(age)
median = np.median(age)
mode = st.mode(age)
sd = np.std(age)
print("Mean age of stars: %.3f"%mean)
print("Median age of stars: %.3f"%median)
print("Mode of stellar age: %.3f"%mode)
print("Standard Deviation: %.3f"%sd)
```

### Plot of Stellar Age

```python
age_sun = 4.5

plt.plot(age,'o',markersize=2)
plt.axhline(mean,color='r',linestyle='--',label='Mean')
plt.axhline(age_sun,color='y',linestyle='--',label='Age of Sun')
plt.axhline(mean+sd,color='k',linestyle='-.',label=r'1$\sigma$')
plt.axhline(mean-sd,color='k',linestyle='-.')
plt.axhline(mean+(2*sd),color='g',linestyle='-.',label=r'2$\sigma$')
plt.axhline(mean-(2*sd),color='g',linestyle='-.')
plt.ylabel('Stellar Age (Gyr)')
plt.legend(loc='best')
plt.show()
```

### Histogram of Stellar Age

```python
plt.hist(age,bins=50,edgecolor='k')
plt.xlabel('Age of host stars')
plt.ylabel('Frequency')
plt.show()
```

- From the above histogram, we can see there is a lot of variance in the age of host stars. Here also, the mean age of host stars is close to the age of Sun, but there are also a lot of stars whose age is different than that of the sun. Also, the histogram has many local peaks which shows that the data is not following any specific kind of distribution but can be thought of as a superposition of many individual distributions.
