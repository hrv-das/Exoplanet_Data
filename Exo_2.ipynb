### Importing Modules and importing data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir(r'/home/hrv/Downloads')
df = pd.read_csv('exodata.csv')
```

### Plot T$_{eff}$ vs log(g) for stars with 3500 ≤ T$_{eff}$ ≤ 6500

```python
df1 = df.dropna(subset = ['st_logg','st_teff'])
teff = np.array([np.array(df1.st_teff)[i] for i in range(len(df1)) if 3500 <= np.array(df1.st_teff)[i] <= 6500])
logg = np.array([np.array(df1.st_logg)[i] for i in range(len(df1)) if 3500 <= np.array(df1.st_teff)[i] <= 6500])

plt.scatter(logg, teff, marker = '*', color = 'orange', edgecolors = 'r',alpha = 0.6)
plt.xlabel(r'log(g) [log10(cm/s$^2$)]')
plt.ylabel("Effective Temperature (K)")
```

- From the above plot, we can see that most stars show a negative correlation between their effective temperature and log(g) values but a few stars show a positive correlation. This means that for most planet hosting stars, lower the g value, higher is its effective temperature. 

### Planet mass vs stellar metallicity[Fe/H]

```python
df2 = df.dropna(subset = ['st_met','pl_bmasse'])
mass = np.array(df2.pl_bmasse)
met = np.array(df2.st_met)

plt.scatter(met, mass, marker = '*', edgecolors = 'darkblue',alpha = 0.6)
plt.xlabel("Stellar Metallicity [Fe/H]")
plt.ylabel("Planet Mass [Earth Mass]")
```

- From the above plot, we can see that most exoplanets have a mass in the range 0.5-5000 times the Earth mass, with only a few of them having very high masses. There is also no evidently visible correlation between the mass of the exoplanet and the metallicity of stars hosting them. But we can see that most of the planet-hosting stars have metallicity in the range -0.6 to 0.6 

### Plot of Planet Radius vs Planet Mass 

```python
df3 = df.dropna(subset = ['pl_rade','pl_bmasse'])
mass = np.array(df3.pl_bmasse)
rad = np.array(df3.pl_rade)


plt.scatter(mass, rad, marker = '*', c = 'lightgreen', edgecolors = 'darkgreen',alpha = 0.6)
plt.xlabel("Planet Mass [Earth Mass]")
plt.ylabel("Planet Radius [Earth Radius]")
```

- From the above plot, we can see that most of the planets do not show any correlation between their mass and radius but there are few planets which show a strong positive correlation between their mass and radius.
